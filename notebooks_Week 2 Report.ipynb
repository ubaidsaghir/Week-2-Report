```
2.Regional Averages: "Find the Average Unit Price of items sold in each region. Which region has the highest average?"
```

```
CREATE TABLE sales_transactions (
trans_id INT PRIMARY KEY,
category VARCHAR(50),
region VARCHAR(50),
quantity INT,
unit_price DECIMAL(10,2),
sale_date DATE
);


INSERT INTO sales_transactions (trans_id,category,region,quantity,unit_price,sale_date) VALUES
(501,'Electronics','North',2,500.00,'2024-03-01'),
(502,'Furniture','South',1,150.00,'2024-03-01'),
(503,'Electronics','North',1,1000.00,'2024-03-02'),
(504,'Furniture','North',3,100.00,'2024-03-02'),
(505,'Electronics','South',1,50.00,'2024-03-03');


SELECT region, ROUND(AVG(unit_price),2) AS average_unit_price
FROM sales_transactions
GROUP BY region
ORDER BY average_unit_price DESC
LIMIT 1;
```

```
CREATE TABLE customers (
customer_id INT PRIMARY KEY,
first_name VARCHAR(50),
last_name VARCHAR(50),
city VARCHAR(50)
);


INSERT INTO customers (customer_id, first_name, last_name, city) VALUES
(10,'Alice','Johnson','New York'),
(11,'Bob','Smith','San Francisco'),
(12,'Charlie','Brown','Austin'),
(13,'David','Miller','New York'),
(14,'Eva','Williams','Los Angeles');


CREATE TABLE orders (
order_id INT PRIMARY KEY,
customer_id INT,
order_date DATE,
total_amount DECIMAL(10,2),
status VARCHAR(50),
FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);


INSERT INTO orders (order_id,customer_id,order_date,total_amount,status) VALUES
(1001,10,'2024-02-01',1200.00,'Shipped'),
(1002,10,'2024-02-15',45.00,'Pending'),
(1003,12,'2024-03-01',300.00,'Shipped');

SELECT * FROM customers;
SELECT * FROM orders;


SELECT c.first_name,o.order_id,o.total_amount
FROM orders o
INNER JOIN customers c
ON o.customer_id = c.customer_id;
```

```
4.The Status Logic: "Join the tables and show the customer's full name and their order status. 
Use a CASE WHEN statement to rename the statuses: 'Shipped' becomes 'On the Way' and 'Pending' becomes 'Processing'."
```

```
CREATE TABLE customers (
customer_id INT PRIMARY KEY,
first_name VARCHAR(50),
last_name VARCHAR(50),
city VARCHAR(50)
);


INSERT INTO customers (customer_id, first_name, last_name, city) VALUES
(10,'Alice','Johnson','New York'),
(11,'Bob','Smith','San Francisco'),
(12,'Charlie','Brown','Austin'),
(13,'David','Miller','New York'),
(14,'Eva','Williams','Los Angeles');


CREATE TABLE orders (
order_id INT PRIMARY KEY,
customer_id INT,
order_date DATE,
total_amount DECIMAL(10,2),
status VARCHAR(50),
FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);


INSERT INTO orders (order_id,customer_id,order_date,total_amount,status) VALUES
(1001,10,'2024-02-01',1200.00,'Shipped'),
(1002,10,'2024-02-15',45.00,'Pending'),
(1003,12,'2024-03-01',300.00,'Shipped');

SELECT * FROM customers;
SELECT * FROM orders;


SELECT CONCAT_WS(' ',c.first_name,c.last_name) AS full_name,
CASE
    WHEN o.status = 'Shipped' THEN 'On the way'
	WHEN o.status = 'Pending' THEN 'Processing'
END AS order_status
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id;
```
